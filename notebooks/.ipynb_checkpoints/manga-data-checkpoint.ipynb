```python
def make_map(dap, name):
    mm= np.zeros(dap.xpos.shape)
    ibin= np.nonzero(dap.binid >= 0)
    mm[ibin]= getattr(dap, name)[dap.binid[ibin]]
    mm=mm.reshape(dap.sqrt_n_spaxels, dap.sqrt_n_spaxels)
    return mm
```

```python
dap.stvel= dap.stvel-np.median(dap.stvel)
```

```python
names= np.array(['stvel', 'stvdisp', 'Fe5270_5335', 'D4000', 'halpha_fb', 'oiii5007_fb'])
labels= np.array([r'$v_\ast$ (km/s)', r'$\sigma_\ast$ (km/s)', r'$\langle {\rm Fe}\rangle$ ($\AA$)', 
                  r'D$_{\rm n}$4000', r'H$\alpha$ flux', r'[OIII] flux'])
vmaxs= np.array([111., 100., 2.91, 2.2, 8., 0.6])
vmins= np.array([-111., 5., 1.2, 1.0, 0., 0.])
```

```python
insa= np.nonzero(nsa['NSAID'] == nsaid)[0]
```

```python
cutoutfile= http_nsa.get('cutout_jpg', nsaid=nsaid, version='v1_0_0')
```

```python
def hexpoints(xcen, ycen, side):
    x= np.zeros(7)
    y= np.zeros(7)
    y[0]=0.
    x[0]=-side
    y[1]=side*np.sqrt(3.)/2.
    x[1]=-1./2.*side
    y[2]=y[1]
    x[2]=-x[1]
    y[3:6]=-y[0:3]
    x[3:6]=-x[0:3]
    x[6]=x[0]
    y[6]=y[0]
    return (xcen+x, ycen+y)
```

```python
fig=plt.figure(dpi=300, figsize=(6.5, 5.))
yox= 5./6.5
imsize_actual= 0.37
imsize= 0.4
mapsize= 0.164
cbarsize= mapsize*0.12
pad=0.1
ymapoff=0.0

#ax_full= fig.add_axes([pad, pad, 1.-pad, 1.-2.*pad])
ax_image= fig.add_axes([pad, 1.-pad-imsize_actual/yox, imsize_actual, imsize_actual/yox])
ax_image.xaxis.set_ticklabels([])
ax_image.yaxis.set_ticklabels([])
ax_spectro= fig.add_axes([pad, pad, 1.-pad, 1.-2.*pad-imsize/yox])

ax_maps=[]
ax_cbars=[]
for i in range(3):
    for j in range(2):
        xmap=pad+imsize+i*mapsize+mapsize*0.02
        ymap=1.-pad-imsize/yox+j*mapsize/yox+mapsize*0.4/yox+ymapoff
        ax_map_tmp= fig.add_axes([xmap, ymap, mapsize, mapsize/yox])
        ax_map_tmp.xaxis.set_ticklabels([])
        ax_map_tmp.yaxis.set_ticklabels([])
        if j == 0:
            labelbottom=True
            labeltop=False
            ycbar=ymap-cbarsize/yox
            lposition='bottom'
            tposition=0.3
        else:
            labelbottom=False
            labeltop=True
            ycbar=ymap+mapsize/yox
            lposition='top'
            tposition=0.6
        ax_cbar_tmp= fig.add_axes([xmap, ycbar, 
                                  mapsize, cbarsize/yox])
        ax_cbar_tmp.yaxis.set_ticklabels([])
        ax_cbar_tmp.yaxis.set_ticks([])
        ax_maps.append(ax_map_tmp)
        ax_cbars.append(ax_cbar_tmp)
 
nticks=4
for ax, cbar, name, vmax, vmin, label, indx in zip(ax_maps, ax_cbars, names, vmaxs, vmins, labels, 
                                             range(len(ax_maps))):
    cmm= make_map(dap, name)
    im= ax.imshow(cmm.T, interpolation='nearest', cmap=cm.Greys, vmax=vmax,
                  vmin=vmin, origin='lower')
    vrange= np.floor(np.log10(np.abs(vmax-vmin)))
    nspacing= np.abs(vmax-vmin)/np.float32(nticks)*1.2
    aspacing= 10.**np.int32(vrange-1)*np.int32(nspacing/10.**np.int32(vrange-1))
    cnticks= np.ceil(np.abs(vmax-vmin)/aspacing)
    ticks=vmin+(np.arange(cnticks)+0.5)*aspacing
    if (indx % 2) == 0:
        labelbottom=True
        labeltop=False
        ycbar=ymap-cbarsize/yox
        lposition='bottom'
        tposition=0.3
        labelpad=1
    else:
        labelbottom=False
        labeltop=True
        ycbar=ymap+mapsize/yox
        lposition='top'
        tposition=0.6
        labelpad=2
    cbarbar= plt.colorbar(im, cax=cbar, orientation='horizontal')
    cbarbar.set_label(label, fontdict={'fontsize':8})
    cbarbar.set_ticks(ticks)
    cbar.xaxis.set_tick_params(labelbottom=labelbottom, labeltop=labeltop, labelsize=8, pad=2)
    #cbar.xaxis.set_ticklabels(str(ticks), position=(0,tposition))
    cbar.xaxis.labelpad=labelpad
    cbar.xaxis.set_label_position(lposition)
    cbarbar.changed()

import matplotlib.image as mpimg
a=mpimg.imread(cutoutfile)
xcen= a.shape[0]*0.5
ycen= a.shape[1]*0.5
side= side_length/0.396
(x,y)= hexpoints(xcen, ycen, side)
ax_image.imshow(a, interpolation='nearest')
ax_image.plot(x, y, color='white', linewidth=0.75)
cubesize=cmm.shape[0]*0.5/0.396
ax_image.set_xlim(xcen-0.5*cubesize, xcen+0.5*cubesize)
ax_image.set_ylim(xcen+0.5*cubesize, xcen-0.5*cubesize)
ax_image.xaxis.set_ticklabels([])
ax_image.yaxis.set_ticklabels([])
ax_image.set_title(str(plate)+'-'+str(ifudesign)+'   NSAID='+str(nsa['NSAID'][insa][0]),
                   fontdict={'fontsize':10})
    
indx= np.arange(10)
for i in indx:
    vdisp=150.
    pixsize= np.log(10.)*2.99792e+5*np.float32(0.0001)
    sigma= vdisp/pixsize
    sflux= filters.gaussian_filter1d(radial.galaxy[i,:], sigma=sigma)
    ax_spectro.plot(radial.wave_obs, np.log10(sflux), color=str(0.7*(i+1)/10.),
                   linewidth=i*0.07+0.6, zorder=i)
ax_spectro.set_ylim(-1.5, 0.30)
ax_spectro.set_xlim(3500., 10450.)
ax_spectro.set_ylabel(r'log$_{10}$ $f_\lambda$ ($10^{-17}$ erg cm$^{-2}$ s$^{-1}$ $\AA^{-1}$)', fontsize=8)
ax_spectro.set_xlabel(r'Wavelength ($\AA$)', fontsize=8)

fig.savefig(os.path.join('/Users', 'blanton', 'sdss4-overview', 'tex', 'figures', 'manga-data.pdf'))
```

```python
width = 5.5
height = 5.5

fig=plt.figure(dpi=300, figsize=(width, height))

pad = 0.12
rpad = 0.05
image_width = 0.5 * (1. - pad - rpad)
image_height = image_width / height * width
map_width = image_width
map_height = image_height

spectrum_width = 1. - pad - rpad
spectrum_height = 1. - image_height - 2. * pad
cbarsize = map_height * 0.1

ax_image = fig.add_axes([pad, 1. - pad - image_height, image_width, image_height])
ax_image.xaxis.set_ticklabels([])
ax_image.yaxis.set_ticklabels([])

ax_map = fig.add_axes([pad + image_width, 1. - pad - image_height, map_width, map_height])
ax_map.xaxis.set_ticklabels([])
ax_map.yaxis.set_ticklabels([])

ax_spectro = fig.add_axes([pad, pad, spectrum_width, spectrum_height])

a=mpimg.imread(cutoutfile)
xcen= a.shape[0]*0.5
ycen= a.shape[1]*0.5
side= side_length/0.396
(x,y)= hexpoints(xcen, ycen, side)
ax_image.imshow(a, interpolation='nearest')
ax_image.plot(x, y, color='white', linewidth=0.75)
cubesize=cmm.shape[0]*0.5/0.396
ax_image.set_xlim(xcen-0.5*cubesize, xcen+0.5*cubesize)
ax_image.set_ylim(xcen+0.5*cubesize, xcen-0.5*cubesize)
ax_image.xaxis.set_ticklabels([])
ax_image.yaxis.set_ticklabels([])
ax_image.set_title(str(plate)+'-'+str(ifudesign)+'   NSAID='+str(nsa['NSAID'][insa][0]),
                   fontdict={'fontsize':10})

cmm= make_map(dap, 'halpha_fb')
vmax = 80.
vmin = 0.
im= ax_map.imshow(cmm.T, interpolation='nearest', cmap=cm.Greys, vmax=vmax,
              vmin=vmin, origin='lower')
vrange= np.floor(np.log10(np.abs(vmax-vmin)))
nspacing= np.abs(vmax-vmin)/np.float32(nticks)*1.2
aspacing= 10.**np.int32(vrange-1)*np.int32(nspacing/10.**np.int32(vrange-1))
cnticks= np.ceil(np.abs(vmax-vmin)/aspacing)
ticks=vmin+(np.arange(cnticks)+0.5)*aspacing
labelbottom=False
labeltop=True
ycbar= 1. - pad
lposition='top'
tposition=0.6
labelpad=2
ax_cbar= fig.add_axes([pad + image_width, ycbar, map_width, cbarsize])
ax_cbar.yaxis.set_ticklabels([])
ax_cbar.yaxis.set_ticks([])

cbarbar= plt.colorbar(im, cax=ax_cbar, orientation='horizontal')
cbarbar.set_label(r'H$\alpha$ flux', fontdict={'fontsize':10})
cbarbar.set_ticks(ticks)
ax_cbar.xaxis.set_tick_params(labelbottom=labelbottom, labeltop=labeltop, labelsize=8, pad=2)
#cbar.xaxis.set_ticklabels(str(ticks), position=(0,tposition))
ax_cbar.xaxis.labelpad=labelpad
ax_cbar.xaxis.set_label_position(lposition)
cbarbar.changed()

indx= [0, 3, 5, 7]
for i in indx:
    vdisp=150.
    pixsize= np.log(10.)*2.99792e+5*np.float32(0.0001)
    sigma= vdisp/pixsize
    sflux= filters.gaussian_filter1d(radial.galaxy[i,:], sigma=sigma)
    ax_spectro.plot(radial.wave_obs, np.log10(sflux), color=str(0.7*(i+1)/10.),
                   linewidth=i*0.07+0.6, zorder=i)
ax_spectro.set_ylim(-0.8, 1.50)
ax_spectro.set_xlim(3500., 10450.)
ax_spectro.set_ylabel(r'log$_{10}$ $f_\lambda$ ($10^{-17}$ erg cm$^{-2}$ s$^{-1}$ $\AA^{-1}$)', fontsize=8)
ax_spectro.set_xlabel(r'Wavelength ($\AA$)', fontsize=10)

filename=os.path.join('/Users', 'blanton', 'Desktop', 'manga-fig-'+str(plate)+'-'+str(ifudesign)+'.eps')
fig.savefig(filename, format='eps')
```
