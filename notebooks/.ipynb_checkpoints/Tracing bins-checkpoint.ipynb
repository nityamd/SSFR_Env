```python
def niceHist(ax, edges, counts, *args, **kwargs):

    N = counts.shape[0]
    X = np.empty(2*N)
    Y = np.empty(2*N)

    X[::2] = edges[:-1]
    X[1::2] = edges[1:]
    Y[::2] = counts[:]
    Y[1::2] = counts[:]

    ax.plot(X, Y, *args, **kwargs)

if __name__ == "__main__":

    N = 10
    edges = np.linspace(0,10, num = N+1)
    counts = np.random.rand(10)

    fig, ax = plt.subplots()
    niceHist(ax, edges, counts, c = 'k')
    plt.xscale('log')
    plt.show()
```
