```python
opt = np.asarray([d['ABSMAG'][:,1][i] - d['ABSMAG'][:,4][i] for i in range(len(d['NSAID']))])
print min(opt), max(opt)
indo1 = np.where(opt<=7.5)[0]
do = d[indo1]
print np.shape(do)
opto = opt[indo1]
indo2 = np.where(opto>=0)[0]
do = do[indo2]
print np.shape(do)
```

```python
delta_m1 = 2.699
delta_m2 = 3.339
delta_m3 = 5.174
delta_m4 = 6.620

inf = np.asarray([do['W1MAG'][i] - do['W3MAG'][i] for i in range(len(do['NSAID']))])
print min(inf), max(inf)
indf1 = np.where(inf>=-3.0)[0]
infi = inf[indf1]
di = do[indf1]
print np.shape(di)
indf2 = np.where(infi<=3.0)[0]
di=di[indf2]
print np.shape(di)

```

```python


ind = di['NSAID']
opt = [di['ABSMAG'][:,1][i] - di['ABSMAG'][:,4][i] for i in range(len(ind))]
inf = [di['W1MAG'][i] - di['W3MAG'][i]  for i in range(len(ind))]

h = binned_statistic_2d(inf,opt,ind,statistic = 'count',bins = (25,25))

yedges = h[1]
xedges = h[2]
xbins = [0.5*(h[1][i] + h[1][i+1]) for i in range(25)]
ybins = [0.5*(h[2][i] + h[2][i+1]) for i in range(25)]
extent = [xbins[0], xbins[-1], ybins[0], ybins[-1]]
numpy.ma.masked_invalid
number = np.ma.array(h[0], mask = h[0] == 0)
number = np.log10(number)

plt.pcolormesh(yedges,xedges,(np.transpose(number)),cmap = plt.cm.binary)
plt.colorbar()
# plt.xlim(0,0.056)
# plt.ylim(min(gr),max(gr))
plt.ylim(0,8.0)
plt.xlim(-3.0,3.0)
plt.xlabel('W1-W3')
plt.ylabel('N-r')
plt.show()
```

```python
# Investigating IR colors;

delta_m1 = 2.699
delta_m2 = 3.339
delta_m3 = 5.174
delta_m4 = 6.620

inf1 = np.asarray([do['W1MAG'][i] - do['W3MAG'][i] for i in range(len(do['NSAID']))])
print min(inf1), max(inf1)
inf2 = np.asarray([do['W1MAG'][i] - do['W3MAG'][i] + (delta_m3 - delta_m1) for i in range(len(do['NSAID']))])
print min(inf2), max(inf2)

z = do['Z']

plt.scatter(z,inf1, s = 0.01)
plt.xlabel('z')
plt.ylabel('[W1]-[W3] (Vega)')
plt.show()
```
